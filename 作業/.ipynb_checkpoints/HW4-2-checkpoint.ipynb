```python
#Step0. 預備動作
bankdict={"004":"台灣銀行","005":"土地銀行","006":"合作金庫",
          "007":"第一銀行","008":"華南銀行","009":"彰化銀行",
          "017":"兆豐商銀","050":"台灣企銀"}
notbankdict={ "011":"上海銀行","012":"富邦銀行","013":"國泰世華",
              "016":"高雄銀行","021":"花旗銀行",
              "039":"澳盛銀行","048":"王道銀行",
              "052":"渣打銀行","053":"台中銀行","081":"匯豐銀行",
              "101":"瑞興銀行","102":"華泰銀行","103":"新光銀行",
              "108":"陽信銀行","147":"三信銀行","803":"聯邦銀行",
              "805":"遠東銀行","806":"元大銀行","807":"永豐銀行",
              "808":"玉山銀行","809":"凱基銀行","810":"星展銀行",
              "812":"台新銀行","815":"日盛銀行","816":"安泰銀行",
              "822":"中國信託"}
totalbankdict={"004":"台灣銀行","005":"土地銀行","006":"合作金庫",
               "007":"第一銀行","008":"華南銀行","009":"彰化銀行",
               "017":"兆豐商銀","050":"台灣企銀",
               "011":"上海銀行","012":"富邦銀行","013":"國泰世華",
               "016":"高雄銀行","021":"花旗銀行",
               "039":"澳盛銀行","048":"王道銀行",
               "052":"渣打銀行","053":"台中銀行","081":"匯豐銀行",
               "101":"瑞興銀行","102":"華泰銀行","103":"新光銀行",
               "108":"陽信銀行","147":"三信銀行","803":"聯邦銀行",
               "805":"遠東銀行","806":"元大銀行","807":"永豐銀行",
               "808":"玉山銀行","809":"凱基銀行","810":"星展銀行",
               "812":"台新銀行","815":"日盛銀行","816":"安泰銀行",
               "822":"中國信託"}
import time

class BankAccount():            
   
    def __init__(self, username, money):   
        self.__name=username               
        self.__balance = money             
        self.title='Taipei First Bank'   

    def get_balance(self):               
        print(self.__name,'的帳戶餘額為: ', self.__balance)

    def save_money(self, money):       
        self.__balance += money
        print("存款", money, "元，完成")        

    def withdraw_money(self, money):    
        self.__balance -= money
        self.__balance -=30
        print("轉帳", money, "元，完成")
        
mybank=BankAccount('您', 5000000)



#Step1. 選擇想要的付款方式+防呆機制
while True:
    payin = input("請輸入想要的付款方式(請輸入數字):"
                  "\n1.轉帳\n2.信用卡\n")
    if payin == "1" or payin == "2":    #選1跟2的話，就跳出迴圈
        break
    else:                               #要不然就一直陷入迴圈(防呆)
        print("你輸入的格式有誤\n")

#Step2. 付款流程+防呆機制
        
#Step2.1 轉帳流程+防呆機制      
if payin == "1":
    while True:
        bankin = input("請輸入您帳戶的銀行代碼:")
        if bankin in totalbankdict:
            #只要是銀行就好(否則就重複陷入迴圈)
            print("你銀行帳戶名稱為",totalbankdict.get(bankin))
            while True:
                cards = input("請輸入你的帳號:\n")
                if cards.isdecimal():
                #帳號格式正確即可，並無實際用途
                    print("讀取中，請稍後。\n")
                    time.sleep(2)
                    print("讀取成功。\n") 
                    break
                else:
                    print("請重新輸入你的帳號")
            break
  
#Step2.2 信用卡流程+防呆機制    
else:
    while True:
        bankin = input("請輸入你的信用卡銀行代碼:")
        #只有合作的銀行才可以用
        if bankin in bankdict:
            print("你用的信用卡銀行為",bankdict.get(bankin))
            while True:
                check = input("你確定要用?要:1、不要:0")
                #確認是否要用該銀行的信用卡
                if check == "1":
                    break
                elif check != "0":
                    print("你輸入的格式有誤，請重新輸入謝謝!!\n")
                #跳出該迴圈，然後重新選擇信用卡銀行
                else:
                    break
            if check == "1":
                while True:
                    cards = input("請輸入你的信用卡帳號:\n")
                    #帳號格式正確即可，並無實際用途
                    if cards.isdecimal():
                        print("讀取中，請稍後。\n")
                        time.sleep(2)
                        print("讀取成功。\n") 
                        break
                    else:
                        print("請重新輸入你的信用卡帳號")
                break

        #只有合作的銀行才可以用
        elif bankin in notbankdict :
            print("沒有支援這家銀行的信用卡，請重新輸入謝謝!!\n")
        else:
            print("你輸入的格式有誤，請重新輸入謝謝!!\n")


#Step3. 輸入每人消費金額以及人數+防呆機制            
while True:
    moneyin = input("請輸入個人消費的金額(新台幣):")
    if moneyin.isdecimal():
        moneyin = int(moneyin)
        break
    
while True:
    people = input("請輸入消費人數:")
    if people.isdecimal():
        people = int(people)
        break   




#Step4. 結帳+滿額贈禮
#Step4.1 轉帳流程+防呆機制
#5萬以下為原價
#5~12萬可以加價升級商務艙
#12萬以上免費升級商務艙
if payin == "1":
    if (moneyin > 50000) and (moneyin <= 120000):
        while True:
            upg = input("是否多消費10000元，將機票升級為商務艙?(要:1、不要:0)")
            #防呆機制(只有要跟不要兩種選擇，要者再另外+1萬)
            if upg == "0" or upg == "1":
                break
            if upg == "1":
                moneyin += 10000
 
    elif moneyin > 120000:
        while True:
            busi_class = input("您的機票艙等是否為經濟艙?(是:1、否:0)")
            #防呆機制(只有是跟不是兩種選擇，是者免費升級，不是者補助一下)
            if busi_class == "0" or busi_class == "1":
                break
        if busi_class == "0":
            moneyin -= 15000
        else:
            print("我們已將您的機票升級為商務艙\n")
#Step4.2 轉帳流程+防呆機制
#3.5萬以下為原價
#3.5~10萬可以加價升級商務艙
#10萬以上免費升級商務艙
else:
    if (moneyin > 35000) and (moneyin <= 100000):
        while True:
            upg = input("是否多消費5000元，將機票升級為商務艙?(要:1、不要:0)")
            #防呆機制(只有要跟不要兩種選擇，要者再另外+5000)
            if upg == "0" or upg == "1":
                break
            if upg == "1":
                moneyin += 5000
    elif moneyin > 100000:
        while True:
            busi_class = input("您的機票艙等是否為經濟艙?(是:1、否:0)")
            #防呆機制(只有是跟不是兩種選擇，是者免費升級，不是者補助一下)
            if busi_class == "0" or busi_class == "1":
                break
        if busi_class == "0":
            moneyin -= 10000
        else:
            print("我們已將您的機票升級為商務艙")

#Step5.交易處理
#轉帳消費部分，滿7人打9折
#信用卡消費部分，滿5人打8折
total = moneyin * people
if (people >= 7) and (payin == "1"):
    total*=0.9
elif (people>=5) and (payin == "2"):
    total*=0.8

print("\n本次消費金額總共為",total,"元\n")
time.sleep(2)
print("交易處理中，請稍後。\n")
time.sleep(2)
print("交易完成\n")
time.sleep(1)
if payin == "1":
    mybank.withdraw_money(total)
    mybank.get_balance()
print("謝謝你的惠顧\n")
```
