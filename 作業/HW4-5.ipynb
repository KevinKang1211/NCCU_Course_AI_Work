```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("death toll.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(7,15))

labels = df["Country"]
separeted = (0, 0, 0, 0, 0.3, 0, 0, 0, 0, 0)
size = df["Deaths"]

plt.pie(size, 
        labels = labels,
        autopct = "%1.1f%%",
        explode = separeted,
        pctdistance = 0.8,
        textprops = {"fontsize": 10})

plt.axis('equal')
plt.title("COVID-19 Death Toll by Country(Top10)")
plt.legend(loc = "best")
plt.show()
```
