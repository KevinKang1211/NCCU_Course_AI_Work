```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("number of car accident.csv")
```

```python
df.head(3)
```

```python
plt.figure(figsize=(6,9))

labels = df["Reason Of Car Accident"]
separeted = (0, 0.4, 0, 0, 0)
size = df["Count"]

plt.pie(size, 
        labels = labels,
        autopct = "%1.1f%%",
        explode = separeted,
        pctdistance = 0.8,
        textprops = {"fontsize": 10})

plt.axis('equal')
plt.title("Pie Chart of Car Accident")
plt.legend(loc = "best")
plt.show()
```
